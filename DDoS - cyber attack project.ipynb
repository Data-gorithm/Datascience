```python
corr_matrix = x.corr()
```

```python
threshold = 0.6

correlation_col = set ()
```

```python
for row in range(len(corr_matrix.columns)):
    for col in range(row):
        if abs(corr_matrix.iloc[row][col]) > threshold:
            col = corr_matrix.columns[row]
            correlation_col.add(corr_matrix.columns[row])
            print(col)
            
```

```python
correlation_col           
```

```python
x.drop(correlation_col, axis = 1, inplace = True)
```
